```python
import pandas as pd
from pytrends.request import TrendReq
```

```python
pytrend = TrendReq()
```

# Interest by region
Let us see the terms which are popular by region worldwide. Search term is "anal sex".

```python
search_terms = ['Bastian Schweinsteiger']
pytrend.build_payload(kw_list=search_terms)

# interest by region
df = pytrend.interest_by_region()
df.head(10)
```

```python
df.reset_index().plot(x='geoName', y='Bastian Schweinsteiger', figsize=(120, 10), kind ='bar')
```

```python
# Get Google's hot trends 
df_hot_ger = pytrend.trending_searches(pn='germany')
df_hot_ger.head(20)
```

# Related queries
Return related queries to the topic "Macaulay Culkin".

```python
pytrend.build_payload(kw_list=['Macaulay Culkin'])
```

```python
# related queries, returs a dictionary of dataframes
related_queries = pytrend.related_queries()
related_queries.values()
```

# Related topics

```python
# related topics, returns a dicitionary of dataframes
pytrend.build_payload(kw_list=['Macaulay Culkin'])
related_topic = pytrend.related_topics()
related_topic.values()
```
