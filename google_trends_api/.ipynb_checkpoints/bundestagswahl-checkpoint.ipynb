```python
import pandas as pd
from pandas_profiling import ProfileReport
from datetime import date
from pytrends.request import TrendReq
from matplotlib import pyplot as plt, figure
%matplotlib inline 
```

```python
pytrend = TrendReq()
```

```python
# provide a set of search terms
kw_list = ['Annalena Baerbock'
           , 'Armin Laschet'
           , 'Markus Söder'
           , 'Olaf Scholz'
           , 'Christian Lindner']
```

```python
# Build payload
pytrend.build_payload(kw_list
                      , timeframe=f'2021-01-01 {date.today()}'
                      , geo='DE')
```

```python
## API methods

# interest over time
df = pytrend.interest_over_time()
df.head(52)
```

```python
# Data prep. with pandas profiling
profile = ProfileReport(df, title="Pandas Profiling Report")
profile.to_widgets()

# store pandas profiling report
# profile.to_file("pd_profiling_report.html")
```

```python
fig = plt.figure(figsize=(16, 6))

ax = plt.gca()

df.plot(kind='line'
        , use_index=True
        , y='Annalena Baerbock'
        , color='seagreen'
        , ax=ax
        , marker='o'
        , alpha=0.4)
df.plot(kind='line'
        , use_index=True
        , y='Armin Laschet'
        , color='dimgray'
        , ax=ax
        , marker='o'
        , alpha=0.4)
df.plot(kind='line'
        , use_index=True
        , y='Markus Söder'
        , color='lightgray'
        , ax=ax
        , marker='o'
        , alpha=0.4)
df.plot(kind='line'
        , use_index=True
        , y='Olaf Scholz'
        , color='firebrick'
        , ax=ax
        , marker='o'
        , alpha=0.4)
df.plot(kind='line'
        , use_index=True
        , y='Christian Lindner'
        , color='gold'
        , ax=ax
        , marker='o' 
        , alpha=0.4)

plt.axvline('2021-04-18', linestyle='dashed', color='palevioletred', alpha=0.6)
plt.text('2021-04-10', 95, r'Baerbock Bekanntgabe Kandidatur', fontsize=10,color='firebrick')

plt.axvline('2021-09-26', linestyle='dashed', color='palevioletred', alpha=0.6)
plt.text('2021-09-06', 95, r'Bundestagswahl', fontsize=10,color='firebrick')

plt.axvline('2021-12-05', linestyle='dashed', color='palevioletred', alpha=0.6)
plt.text('2021-10-27', 95, r'Konstituierende Sitzung', fontsize=10,color='firebrick')

plt.gca().legend(('Baerbock (Grüne)','Laschet (Union)','Söder (Union)','Scholz (SPD)','Lindner (FDP)'))
plt.title("Google trends - Bundestagswahl")

plt.show()
```
